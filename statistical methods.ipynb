```python
# import used libraries
import pandas as pd
import numpy as np
from sklearn.datasets import load_boston
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
import copy
```

```python
# feature scalling (standardization)
def featureScalling (X):
  scaler = StandardScaler()
  return scaler.fit_transform(X)
```

```python
# calculating the cost
def cost(coff, X, Y):
  coffx = copy.deepcopy(coff)
  xx = copy.deepcopy(X)
  yx = copy.deepcopy(Y)
  if(isinstance(coffx, pd.DataFrame)):
    coffx = coffx.to_numpy()
  if(isinstance(xx, pd.DataFrame)):
    xx = xx.to_numpy()
  if(isinstance(yx, pd.DataFrame)):
    yx = yx.to_numpy()
  part1 = np.matmul(xx, coffx) # x*coff
  part2 = np.subtract(part1, yx) # X*coff - y
  part3 = np.transpose(part2) # transpose of x * coff - y
  result = np.matmul(part3, part2)
  return result.sum()
```

```python
# calculatin R-square
def rsquare(X, Y):
  xx = copy.deepcopy(X)
  yx = copy.deepcopy(Y)
  j, coff = gradientAlgo(xx, yx)
  if(isinstance(yx, pd.DataFrame)):
    yx = yx.to_numpy()
  mean = yx.sum()
  mean = mean / yx.size
  sst = 0
  for i in range(0, yx.size):
    sst = sst + pow((yx[i][0] - mean), 2)
  result = (sst - j)/sst
  return result
```

```python
# load the boston dataset
boston_data = load_boston()
df_boston = pd.DataFrame(boston_data.data,columns=boston_data.feature_names)
df_boston['target'] = pd.Series(boston_data.target)
df_boston.head()
```

```python
# printing the featueres 
X = df_boston.drop('target',axis = 1)
X
```

```python
# displaying the output to be prediected thorug the dataset
dtarget = df_boston['target']
target = pd.DataFrame(dtarget)
target
```

```python
# applying the model
def gradientAlgo(X, Y):
  xx = X.copy(deep=True)
  yx = Y.copy(deep=True)
  xx = featureScalling(xx)
  xx = pd.DataFrame(xx)
  xx.insert(0,'ones',1)
  xx = xx.to_numpy()
  row_x, col_x = xx.shape
  coff = np.ones((col_x, 1))
  if(isinstance(yx, pd.DataFrame)):
    yx = yx.to_numpy()
  while True:
    part1 = np.matmul(xx, coff)
    part2 = np.subtract(part1, yx)
    part3 = np.matmul(np.matrix.transpose(xx), part2)
    coff = np.subtract(coff, 0.0002*part3)
    if(abs(part3.sum()) < 0.002):
      break
  j = cost(coff, xx, yx)
  return j, coff
```

```python
# test the model
test, coff = gradientAlgo(X, target)
print(test)
print(coff)
print(rsquare(X, target))
```

```python
# step forward feature selection
def stepForward(X, Y, tol):
  xx = copy.deepcopy(X)
  yx = copy.deepcopy(Y)
  col_names = xx.columns
  x_columns = []
  for i in range(0, len(col_names)):
    x_columns.append(pd.DataFrame(xx[col_names[i]]))
  result = pd.DataFrame()
  max_index = -1
  max_r = -1
  flag = True
  while len(x_columns) != 0 :
    for i in range(0, len(x_columns)):
      curr_data = pd.concat([result, x_columns[i]], axis=1)
      curr_r = rsquare(curr_data, yx)
      if(curr_r > max_r):
        max_r = curr_r
        max_index = i
    if flag:
      result = pd.concat([result, x_columns.pop(max_index)], axis=1)
      flag = False
    else :
      rr = rsquare(result, yx)
      if abs(rr-max_r) < tol :
        break
      else :
        result = pd.concat([result, x_columns.pop(max_index)], axis=1)
        max_index = -1
        max_r = -1
  return result
```

```python
# step backward feature selection
def stepBackward(X, Y, tol):
  xx = copy.deepcopy(X)
  yx = copy.deepcopy(Y)
  col_names = xx.columns
  col_names = col_names.tolist()
 # print(col_names)
  max_index = -1
  max_r = -1
  flag = True
  while len(col_names) != 0 :
    for i in range(0, len(col_names)):
      curr_data = xx.drop(col_names[i], axis=1)
      curr_r = rsquare(curr_data, yx)
      if(curr_r > max_r):
        max_r = curr_r
        max_index = i
    rr = rsquare(xx, yx)
    if abs(rr - max_r) < tol :
      xx = xx.drop(col_names.pop(max_index), axis=1)
      max_index = -1
      max_r = -1
    else :
      break
  return xx
```

```python
# calculating the predicted values
def predected (coff, x):
  if(isinstance(coff, pd.DataFrame)):
    coff = coff.to_numpy()
  if(isinstance(x, pd.DataFrame)):
    x = x.to_numpy()
  exponent = np.matmul(x, coff)
  exponent_row, exponent_col = exponent.shape
  for i in range(0,exponent_row):
    exponent[i][0] = 1/(1 + np.exp(-1*exponent[i][0]))
  return exponent
```

```python
# calculating the cost
def cost (coff, x, y):
  hx = predected(coff, x)
  result = 0
  row, col = y.shape
  for i in range(0, row):
    result = result - y[i][0]*np.log(hx[i][0]) - (1 - y[i][0])*np.log(1 - hx[i][0])
  return result
```
